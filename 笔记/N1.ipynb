# $\lambda$-return 

Keywords: offline

1. Motivation：从n-step TD中的灵活性和误差减少性质中，更进一步提出平均衰减n-step returns，以更好的降低误差并提升鲁棒性
2. 特点：前向视图(forward view)
3. 复合更新(Compound update)：通过引入衰减参数$\lambda$将不同远近的n-step return线性加权组合成λ-returns，即$G^{\lambda}_t$，然后令$U_t = G^{\lambda}_t$即可
   




[从强化学习和最优化视角理解资格迹-TD(λ)](https://zhuanlan.zhihu.com/p/347529908)

$\lambda$ 在 $\lambda$-return 中是前向视图系数, 在 $TD(\lambda)$ 中是后向视图系数.

资格迹向量 $z_t$ trace的含义: 追踪并记录历史值函数梯度. eligibility 的含义: 权重权重 $\omega$ 中的每个分量有多少资格接受对应的更新.

迹衰减参数和$(\gamma\lambda)$的意义:
1. 对历史值函数梯度进行折扣,从而达到梯度短期记忆的效果, 即重视当前梯度, 淡化历史梯度;
2. 通过$[0, 1]$来调节记忆量.
